```python
def equazioneSecondoGrado(x):
    y = []
    for number in x:
        tmp = (number*number)+(2*number)+2 
        y.append(tmp)
    return y
```

```python
x = [1, 2, 3, 4, 5]
y = equazioneSecondoGrado(x)
```

```python
plt.title("Equazione secondo grado")
plt.plot(x, y)
```

```python
# Plot un cerchio

import matplotlib.pyplot as plt
import numpy as np 

def circle(r, phi):
    return r*np.cos(phi), r*np.sin(phi)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,aspect = 'equal')

phis = np.arange(0, 6.28, 0.01)
r = 19
ax.plot(*circle(r, phis), c='r', ls='-')
plt.show()
```
