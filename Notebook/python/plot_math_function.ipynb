```python
import numpy as np
from matplotlib import pyplot as plt

def PolyCoefficients(x, coeffs):
    """ Returns a polynomial for ``x`` values for the ``coeffs`` provided.

    The coefficients must be in ascending order (``x**0`` to ``x**o``).
    """
    o = len(coeffs)
    print(f'# This is a polynomial of order {o}.')
    y = 0
    for i in range(o):
        y += coeffs[i]*x**i
    return y

x = np.linspace(-10, 10)
coeffs = [1, 2, 3, 4, 5]
plt.plot(x, PolyCoefficients(x, coeffs))
plt.show()
```

```python
from sympy import *
import numpy as np
x = Symbol('x')
coeffs = [1, 2, 3, 4, 5]

def PolyCoefficients(x, coeffs):
    """ Returns a polynomial for ``x`` values for the ``coeffs`` provided.

    The coefficients must be in ascending order (``x**0`` to ``x**o``).
    """
    o = len(coeffs)
    print(f'# This is a polynomial of order {o}.')
    y = 0
    for i in range(o):
        y += coeffs[i]*x**i
    return y

y = PolyCoefficients(x, coeffs)
yprime = y.diff(x)
ysecnd = yprime.diff(x)
print(yprime)
print(ysecnd)
```

```python
x = np.linspace(-10, 10)
plt.plot(x, PolyCoefficients(x, coeffs), 'red')
plt.plot(x, eval(str(yprime)), 'yellow')
plt.plot(x, eval(str(ysecnd)), 'blue')
plt.legend(['func', 'dfunc', 'dfunc^2']) 
plt.show()
```
